```python
def clean_tokenize(s):
    words = RegexpTokenizer(r'\w+').tokenize(s)
    return {word for word in words if word not in stopwords.words("english")}
```

```python
print(list(clean_tokenize(train.question1[0])))
print(train.question1[0])
```
